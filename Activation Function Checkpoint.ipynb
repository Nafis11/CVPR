```python
import matplotlib.pyplot as plt
import numpy as np

def elu(x, alpha=1.0):
    """Exponential Linear Unit (ELU) activation function."""
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Generate input values from -5 to 5
x = np.linspace(-5, 5, 10)

# Calculate the output values of the ELU activation function
y = elu(x)

# Plot the ELU activation function
plt.plot(x, y)
plt.title("Exponential Linear Unit (ELU) Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def selu(x, alpha=1.67326, scale=1.0507):
    
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Generate input values from -5 to 5
x = np.linspace(-5, 5, 100)

# Calculate the output values of the SELU activation function
y = selu(x)

# Plot the SELU activation function
plt.plot(x, y)
plt.title("SELU) Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-10,10)
y=[]
for i in range(x.shape[0]):
    if(x[i]>0):
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y,c="blue")
plt.title(" Step")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-15,15)
y=[]
for i in range(x.shape[0]):
    y.append(1/(1+np.exp(-x[i])))
plt.plot(x,y,c="blue")
plt.title("Sigmoid ")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return t
x=np.arange(-10,10,0.01)
plt.ax=plt.subplots(figsize=(9,5))
plt.plot(x,tanh(x), color="blue", linewidth=3, label="tanh")
plt.legend(loc="upper left", frameon=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    y=max(0,x)
    return y

x=np.arange(-10,10,0.01)
relu_output=[relu(i) for i in x]
plt.ax=plt.subplots(figsize=(9,5))
plt.plot(x,relu_output, color="blue", linewidth=3, label="Relu")
plt.legend(loc="upper left", frameon=False)
plt.show()
```
